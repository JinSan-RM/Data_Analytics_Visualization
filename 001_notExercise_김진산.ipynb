# I. Data Preprocessing

* '기간'열(Column) 정보

```
DF['기간'].value_counts()
```

* '기간'열(Column) <font color = 'red'>삭제

```
DF.drop(['기간'],axis = 1, inplace = True)    
```

* <font color = 'red'>삭제</font> 결과 확인

```
DF.head()
```

> ## 3) 0, 23~52번(지역분류) index 삭제

* '대분류' 정보

```
DF['대분류']
```

* 23번 ~ 52번 index <font color = 'red'>삭제

```
DF
```

```
DF
```

# II. '성별' 통계

> ## 1) '성별' 데이터 추출

```
sex = DF[DF['대분류'] == '성별']
```

```
sex
```

> ## 2) '대분류' 열(Column) 삭제

```
sex.drop('분류', axis = 1,inplace=True)
```

```
sex
```

> ## 3) '분류'열(Column)을 index로 지정

```
sex.rename(columns={'대분류':'분류'}, inplace = True)
sex.set_index('분류',inplace=True)
sex.index = ['남자','여자']
```

```
sex
```

> ## 4) matplotlib 시각화

```
sex['운동을 할 충분한 시간이 없어서']
```

```
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (15, 6))

ax[0, 0].pie(sex['운동을 할 충분한 시간이 없어서'],
          labels = sex['운동을 할 충분한 시간이 없어서'].index,
          autopct='%1.1f%%',
          explode = [0.05, 0.0]
          )
ax[0, 0].set_title('운동을 할 충분한 시간이 없어서')


ax[0, 1].pie(sex['함께 운동을 할 사람이 없어서'],
             labels = sex['함께 운동을 할 사람이 없어서'].index,
             autopct='%1.1f%%',
             explode = [0.05, 0.0]
             )
ax[0, 1].set_title('함께 운동을 할 사람이 없어서')

ax[1, 0].pie(sex['운동을 할 만한 장소가 없어서'],
             labels = sex['운동을 할 만한 장소가 없어서'].index,
             autopct='%1.1f%%',
             explode = [0.05, 0.0])

ax[1, 0].set_title('운동을 할 만한 장소가 없어서')

ax[1, 1].pie(sex['운동을 싫어해서'],
             labels = sex['운동을 싫어해서'].index,
             autopct='%1.1f%%',
             explode = [0.05, 0.0])

ax[1, 1].set_title('운동을 싫어해서')

plt.show()
```

# III. '연령별' 통계

> ## 1) '연령별' 데이터 전처리

```
age = DF[DF['대분류'] == '연령별']
age.drop('대분류', axis=1, inplace= True)
```

```
age.set_index('분류', inplace=True)
```

```
myexplode = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05]
age
```

> ## 2) matplotlib 시각화

```
fig, ax = plt.subplots(ncols=3, figsize = (15,9))

ax[0].pie(age['운동을 할 충분한 시간이 없어서'],
          labels = age['운동을 할 충분한 시간이 없어서'].index,
          autopct='%0.1f%%', explode = myexplode
          )

ax[1].pie(age['함께 운동을 할 사람이 없어서'],
          labels =age['함께 운동을 할 사람이 없어서'].index,
          autopct='%0.1f%%', explode = myexplode
          )
ax[2].pie(age['운동을 할 만한 장소가 없어서'],
          labels =age['운동을 할 만한 장소가 없어서'].index,
          autopct='%0.1f%%', explode = myexplode
          )

ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[2].set_title('운동을 할 만한 장소가 없어서')
plt.show()
```
